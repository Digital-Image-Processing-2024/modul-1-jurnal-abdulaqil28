Script dibawah ini merupakan script untuk menggubah gambar sebelumnya menjadi grayscale. Dengan script cv.COLOR_BGR2GRAY dari gambar yang mode RGB akan menjadi warna Gray. sehingga ketika ditampilkan dengan script plt.imshow, gambar yang muncul adalah gambar yang berwarna gray.

```python
gray_img = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
plt.imshow(gray_img, cmap='gray')
```

```python
def normalisasi(gray_img):
    min_val = gray_img.min()
    max_val = gray_img.max()
    normalisasi_img = (gray_img - min_val) / (max_val - min_val)
    return normalisasi_img
```

Script diatas merupakan fungsi untuk proses normalisasi pada citra digital dengan cara mengubah intensitas piksel dari nilai minimum sampai nilai maksimum dengan rentang nilai 0 hingga 1. Pertama program akan mencari nilai minimum dan maksimum dan disimpan di "min_val" dan "max_val". kemudian setiap nilai piksel dalam citra dikurangi dengan nilai minimum dan dibagi dengan selisih nilai maksimum dan minimum tersebut. sehingga menghasilkan citra yang telah dinormalisasi

```python
normal = normalisasi(gray_img)
print(normal)
```

Script dibawah ini merupakan proses konvolusi citra menggunakan kernel yang disediakan dengan zero padding

kernel:
[1, 2, -1, 1]
[0, 2, 1, 2]
[2, 1, 1, 0]
[1, 0, -1, 0]

konvolusi akan dilakukan dengan cara mengalikan subset citra dengan kernel pada setiap piksel, menjumlahkan hasilnya, dan menyimpan nilai tersebut dalam citra hasil konvolusi. Setelah konvolusi selesai, akan dilakukan normalisasi citra hasil konvolusi agar rentang nilai antara 0 hingga 1. kemudian tampilkan hasilnya menggunakan "plt.show()".

```python
kernel = np.array([
    [1, 2, -1, 1],
    [0, 2, 1, 2],
    [2, 1, 1, 0],
    [1, 0, -1, 0]
])
tinggi_kernel, lebar_kernel = kernel.shape
ukuran_padding = int((tinggi_kernel - 1) / 2)
padding = np.pad(gray_img, ((ukuran_padding, ukuran_padding + 1), (ukuran_padding, ukuran_padding + 1)), mode='constant')
hasil_konvolusi = np.zeros((tinggi, lebar))

for i in range(hasil_konvolusi.shape[0]):
    for j in range(hasil_konvolusi.shape[1]):   
        subset = padding[i:i + tinggi_kernel, j:j + lebar_kernel]
        hasil_perkalian = subset * kernel
        jumlah_perkalian = np.sum(hasil_perkalian)
        hasil_konvolusi[i][j] = np.round(jumlah_perkalian)

# Normalisasi hasil konvolusi
nilai_min = np.min(hasil_konvolusi)
nilai_max = np.max(hasil_konvolusi)
hasil_normalisasi = np.round((hasil_konvolusi - nilai_min) / (nilai_max - nilai_min))

plt.imshow(hasil_normalisasi, cmap='gray')
plt.show()
```

Script dibawah ini merupakan proses melakukan rotasi citra dengan searah jarum jam (90 derajat).Pertama menghitung tinggi dan lebar citra A, lalu membuat citra baru B dengan ukuran lebar dan tinggi yang dibalik dari citra A. Selanjutnya, melakukan perulangan melalui setiap piksel dalam citra A, mengambil nilai piksel dari A dan menyimpannya ke citra B dengan posisi yang sudah dirotasi 90 derajat searah jarum jam. kemudian tampilkan hasilnya menggunakan "plt.show()".

```python
def rotasi(A):
    tinggi, lebar = A.shape
    B = np.zeros((lebar, tinggi), dtype=A.dtype)
    k = lebar - 1
    for i in range(tinggi):
        for j in range(lebar):
            B[j][k] = A[i][j]
        k -= 1
    return B

rotasi_img = rotasi(gray_img)

plt.imshow(rotasi_img, cmap='gray')

plt.show()
```

Script dibawh ini merupakan proses pencerminan terhadap sumbu vertikal (sumbu Y). Pertama, program membuat citra baru  dengan ukuran yang sama seperti citra image. Selanjutnya, dilakukan perulangan melalui setiap piksel dalam citra image. Setiap nilai piksel pada posisi (i, lebar - 1 - j) (pencerminan terhadap sumbu Y) dari citra image dipindahkan ke posisi (i, j) dalam citra flip_img. kemudian tampilkan hasilnya menggunakan "plt.show()". 

```python
def pencerminan(image):
    tinggi, lebar = image.shape
    flip_img = np.zeros((tinggi, lebar), dtype=image.dtype)
    
    for i in range(tinggi):
        for j in range(lebar):
            flip_img[i][lebar - 1 - j] = image[i][j]
    
    return flip_img

sumbuY = pencerminan(rotasi_img)

plt.imshow(sumbuY, cmap='gray')
plt.show()
```

Script dibawah ini merupakan proses dilatasi pada citra image dengan faktor skala yang 3. Pertama, program akan menghitung ukuran baru untuk citra hasil dilatasi berdasarkan faktor skala 3. Selanjutnya, program membuat citra baru dengan ukuran yang baru.  Hal ini dilakukan untuk memperbesar ukuran citra dengan faktor skala yang telah ditentukan. kemudian tampilkan hasilnya menggunakan "plt.show()".

```python
def dilatasi(image, skala=3):
    tinggi, lebar = image.shape
    new_tinggi = tinggi * skala
    new_lebar = lebar * skala
    dilatasi_img = np.zeros((new_tinggi, new_lebar), dtype=image.dtype)
    
    for i in range(new_tinggi):
        for j in range(new_lebar):
            i_awal = i // skala
            j_awal = j // skala
            dilatasi_img[i][j] = image[i_awal][j_awal]
    
    return dilatasi_img

dilated_img = dilatasi(rotasi_img, skala=3)

plt.imshow(dilated_img, cmap='gray')
plt.title("Hasil Dilatasi Manual dengan Skala 3")
plt.show()
```
